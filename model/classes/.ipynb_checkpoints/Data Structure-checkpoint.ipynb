```python
from skimage import io
from skimage.io import imshow
import numpy as np
```

```python
from ZODB import FileStorage, DB
import BTrees.OOBTree
import transaction
import persistent
from datetime import datetime
from datetime import timezone
```

```python
class User(persistent.Persistent):
    #User types: 0-guest, 1-technician, 2-admin
    #Guest: always has default settings, name is guest, no pin
    #technician: can change name, pin, personal settings, can reset to guest settings (aka default)
    #admin: Has all prior permissions & can set default settings
    def __init__(self,userId, name, pin=None, active=True,userType=1, settings=None):
        self.userType = userType
        self.userId = userId
        self.name = 'guest' if(userType == 0) else name
        self.pin = None if (userType == 0) else pin
        self.active = active
        
    def change_pin(self, pin):
        if (userType >= 1):
            self.pin = pin
            self._p_changed = True
        
    def change_type(self, userType):
        #Don't allow users to use this unless admin
        self.userType = userType
        self._p_changed = True
        
    def activate(self):
        self.active = True
        self._p_changed = True
        
    def deactivate(self):
        self.active = False
        self._p_changed = True
        
    def change_name(self, name):
        if (userType >= 1):
            self.name = name
            self._p_changed = True
    
    def change_settings(self,settings):
        if (userType >= 1):
            self.settings=settings
            self._p_changed = True
            
    def change_defaults(self,settings):
        if (userType == 2):
            self.settings=settings
            self._p_changed = True
            
    def get_settings(self):
        if (userType == 0):
            return self.settings
        else:
            print("Failed to pull settings from non-guest user")
            return
```

```python
class Settings:
    def __init__(self, brightness,contrast,sharpness,iso,shutter_speed)
        self.contrast = contrast
        self.sharpness=sharpness
        self.iso=iso
        self.brightness=brightness
        self.shutter_speed = shutter_speed
```

```python
class Result:
    def __init__(self, detected,wavelength=None, strength=None):
        self.detected = detected
        self.wavelength = wavelength
        self.strength = strength
```

```python
class CameraDetails:
    def __init__(self, shutter_speed, awb_gains, brightness, contrast, sharpness, saturation, exposure_comp, iso, meter_mode, resolution):
        self.shutter_speed = shutter_speed
        self.awb_gains = awb_gains
        self.brightness = brightness
        self.contrast = contrast
        self.sharpness = sharpness
        self.saturation = saturation
        self.exposure_comp = exposure_comp
        self.iso = iso
        self.meter_mode = meter_mode
        self.resolution = resolution
        
```

```python
class CaptureDetails:
    def __init__(self, shutter_speed, duration, interval, count):
        self.shutter_speed = shutter_speed
        self.duration = duration
        self.interval = interval
        self.count = count
```

```python
class Trial(persistent.Persistent):
    def __init__(self, trialId,userId, dateTime, notes, scan, result, captureDetails, cameraDetails):
        self.userId = userId
        self.dateTime = dateTime
        self.notes = notes
        self.scan = scan
        self.result = result
        self.captureDetails = captureDetails
        self.cameraDetails = cameraDetails
        
    def edit_notes(self, notes):
        self.notes = notes
        self._p_changed = True
```

```python
from random import randint
def makeCameraDetails():
    return CameraDetails(randint(1,10),randint(1,10),randint(1,10),randint(1,10),randint(1,10),randint(1,10),randint(1,10),randint(1,10),randint(1,10),randint(1,10))
def makeCaptureDetails():
    sp = randint(500,6000)
    duration = int(randint(sp,1200000)/1000) #0 to 1200 seconds (20 mins)
    interval = int(randint(sp,duration*1000)/1000)
    count = duration/interval
    return CaptureDetails(sp,duration,interval,count)
def makeResult():
    return Result(randint(0,1),randint(500,700),randint(0,10))
def makeImageArray(count):
    l = []
    for i in range (count):
        l.append(np.random.rand(1200,1200,3))
    res = np.stack(l)
    return res
def makeSettings():
    return Settings(randint(0,100),randint(0,100),randint(0,1600),randint(0,200000))
def makeUser(userId,name):
    name = names[randint(0,len(names)-1)]
    pin = 1234
    admin = randint(0,1)
    active = True
    settings = makeSettings()
    return User(userId,name,pin,active,admin, settings)
```

```python
def generateUser(userId,name):
    return makeUser(userId,name)

def generateTrial(trialId,userId):
    #userId, dateTime, notes, scan, result, captureDetails, cameraDetails
    return Trial(trialId,userId,datetime.now().astimezone(timezone.utc).timestamp(),str(userId) + ": BLAH BLAH", makeImageArray(randint(1,5)),makeResult(),makeCaptureDetails(),makeCameraDetails())
```

```python
storage = FileStorage.FileStorage('LucereSample.fs')
db = DB(storage)
connection = db.open()
root = connection.root()
```
